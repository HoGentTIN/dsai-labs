The wingspans of 352 Great Tits were measured. The information is summarised in the dataframe in the code block below.

- Visualise the data using a suitable chart type
- Calculate the number of birds using the dataframe
- Calculate the mean of the wingspan using the dataframe
- Calculate the standard deviation of the wingspan using the formula for standard deviation using the dataframe 

$$ s = \sqrt{\sum_{i=1}^{n} {({x_i - \overline{x}})^2} \over n - 1} $$

```python
l = []
birds = pd.DataFrame.from_dict({
    'wingspan':  [70, 71, 72, 73, 74, 75, 76, 77, 78, 79], 
    'frequency': [ 4, 19, 53, 77, 23, 75, 53, 30, 12,  6]})
for n, f in zip(birds['wingspan'], birds['frequency']):
    l.extend([n]*f)
l = np.array(l)
print(birds['frequency'].sum())
print(l.mean())
print(l.std(ddof=1))
```
